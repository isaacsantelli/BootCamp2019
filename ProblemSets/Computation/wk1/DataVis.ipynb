```python
from scipy.stats import linregress
```

```python
for i in range(0,4):
    slope, intercept, r_value, p_value, std_err = linregress(anscombe[:,2*i], anscombe[:,2*i])
    print("Slope is: ", slope)
    print("Intercept is: ", intercept)
    print("Correlation Coefficient is: ", r_value)
```

```python
import scipy
```

```python
x = np.linspace(0, 1, 200)
fig = plt.figure(figsize = [16, 16])
for n in range(0, 4):
    ind = 4*n+1
    for v in range(0, n+1):
        sub = fig.add_subplot(4, 4, ind+v)
        poly = scipy.special.binom(n,v)* x**v * (1-x)**(n-v)
        sub.plot(x, poly, label = "n = {}, v = {}".format(n, v), )
        plt.title("n = {}, v = {}".format(n, v))
        plt.xlim([0, 1.0])
        plt.ylim([0, 1.5])
```

```python
fig = plt.figure(figsize = [15, 5])
htwt = fig.add_subplot(1, 3, 1)
htwt.scatter(mlb[:,0], mlb[:,1], marker='D')
slope, intercept, r_value, p_value, std_err = linregress(mlb[:,0], mlb[:,1])
x = np.linspace(65, 85, 200)
y = intercept + slope * x
htwt.plot(x, y)
plt.title("Height Vs. Weight")
plt.xlabel('Height (in)')
plt.ylabel('Weight (lbs)')
wtage = fig.add_subplot(1, 3, 2)
wtage.scatter(mlb[:,2], mlb[:,1], marker='o', )
plt.title("Age Vs. Weight")
plt.xlabel('Age (years)')
plt.ylabel('Weight (lbs)')
htage = fig.add_subplot(1, 3, 3)
htage.scatter(mlb[:,2], mlb[:,0], marker='*')
plt.title("Age Vs. Height")
plt.xlabel('Age (years)')
plt.ylabel('Height (in)')
plt.ylim(65, 90)
```

```python
x = np.linspace(-1.5, 1.5, 200)
y = np.linspace(-.5, 1.5, 200)
X, Y = np.meshgrid(x, y)
Z = (1 - X)**2 + (100 * (Y - X**2)**2)

fig, ax = plt.subplots(1,2, figsize= [16,8])
htmp = ax[0].pcolormesh(X, Y, Z, cmap="viridis")
fig.colorbar(htmp, ax=ax[0])
cont = ax[1].contour(X, Y, Z, 20, cmap="inferno")
plt.plot(1, 1, marker = 'd', label = "Minimizer")
fig.colorbar(cont, ax=ax[1])
```
