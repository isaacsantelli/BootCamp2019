```python
def randgrid(n):
    '''
    The function takes a number n as an input and creates an
    nxn grid of normally distributed numbers
    '''
    arr = np.random.normal(size =(n,n))
    means = arr.mean(axis = 1)
    return means.var()
```

```python
randgrid(5)
```

```python
def varN():
    '''
    Calls randgrid on various levels of N 
    '''
    L = []
    for i in range(1,11):
        L.append(randgrid(i*100))
    return np.asarray(L)
```

```python
rv = varN()
```

```python
rv
```

```python
x = np.arange(1,11)*100
plt.plot(x, rv)
plt.xlabel("# of Random Draws From Normal")
plt.ylabel("Variance of means of draws")
```

```python
def plotcurve1():
    '''
    Plots the curve 1/(x-1) on the domain [-2,6]
    '''
    x1 = np.linspace(-2, .999, 2000)
    x2 = np.linspace(1.001, 6, 2000)
    plt.plot(x1, 1/(x1-1), 'm--')
    plt.plot(x2, 1/(x2-1), 'm--')
    plt.xlim(-2,6)
    plt.ylim(-6,6)
    return
```

```python
def plotcurves2():
    '''
    function which plots the  the functions sin(x), sin(2x),
    2 sin(x), and 2 sin(2x) on the domain [0, 2π], each in 
    a separate subplot of a single figure.
    '''
    x = np.linspace(0, 2*np.pi, 1000)
    fig, ax = plt.subplots(2,2, figsize = [16, 16])
    fig.suptitle("Different Sine Graphs", fontsize=20)
    plt.title("Some Graphs")
    sinx = plt.subplot(2,2,1)
    plt.plot(x, np.sin(x), 'g-')
    plt.ylim(-2,2)
    plt.xlim(0, 2*np.pi)
    plt.title("Sin(x)")
    sin2x = plt.subplot(2,2,2, sharey=sinx, sharex=sinx)
    plt.plot(x, np.sin(2*x), 'r--')
    plt.title("Sin(2x)")
    sinx_2 = plt.subplot(2,2,3, sharey=sinx, sharex=sinx)
    plt.plot(x, 2*np.sin(x), 'b--')
    plt.title("2Sin(x)")
    sin2x_2 = plt.subplot(2,2,4, sharey=sinx, sharex=sinx)
    plt.plot(x, 2*np.sin(2*x), 'm:')
    plt.title("2Sin(2x)")
    return
```

```python
plotcurves2()
```
